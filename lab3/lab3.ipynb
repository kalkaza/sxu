```python
"""Lab3"""

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#zakres osi x - określenie przyjmowanych wartości
x = np.linspace(0, 5, 100)
print(x)
```

```python
#wykres funkcji liniowej
plt.plot(x, x, 'r', label='Funkcja liniowa')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Wykres funkcji liniowej")

plt.grid(True)
```

```python
#wykres funkcji kwadratowej
plt.plot(x, x**2, 'g', label='Funkcja kwadratowa')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Wykres funkcji kwadratowej")
plt.legend(loc="upper right")
```

```python
#wykres funkcji kwadratowej
x2 = np.linspace(-10,10,100)

plt.plot(x2, x2**2, label='Funkcja kwadratowa')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Wykres funkcji kwadratowej")
plt.legend(loc="upper right")
```

```python
#wykres funkcji  trzeciego rzędu
plt.plot(x, x**3, 'y', label='Funkcja trzeciego rzędu')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Wykres funkcji trzeciego rzędu")
plt.legend(loc="upper right")
```

```python
#wykres funkcji  trzeciego rzędu

plt.plot(x2, x2**3, 'p', label='Funkcja trzeciego rzędu')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Wykres funkcji trzeciego rzędu")
plt.legend(loc="upper right")
```

```python
#wykres funkcji cosinus
x3 = np.linspace(-np.pi,np.pi,100)

y = np.cos(x3)
plt.plot(x3, y, 'or', label='cosinus')
plt.xlabel('x')
plt.ylabel('y')
plt.title("wykres cos")
plt.legend(loc="upper right")
```

```python
#wykres funkcji sinus
y = np.sin(x3)

plt.plot(x3, y, 'gp', label='sinus')
plt.xlabel('x')
plt.ylabel('y')
plt.title("wykres sin")
plt.legend(loc="upper right")
```

```python
#import pliku z autami
data = pd.read_csv("samochody1tys.csv")
data
```

```python
#policz ilości marek a następnie zapisz to do obiektu frame
df = data.marka.value_counts().to_frame()
df
```

```python
 # narysuj na podstawie ilości marek 
df.plot.bar(y='marka')
plt.ylabel('Ilość')
plt.xlabel('Marki')
plt.title('Ilość marek')
```

```python
df = pd.Index(data['wojewodztwo']).value_counts()
df
```

```python
#Wykres kołowy
#pobranie wartości 
labels = df.keys().tolist()
#pobreanie ilości wystąpień
sizes = df.tolist()

#narysowanie wykresu ilości wystąpień i obrócenie go o 90 stopni
patches, texts = plt.pie(sizes, startangle=90)
plt.legend(patches, labels) #utworzenie legendy na podstawie wojewodztwa
plt.axis('equal') # zmiana położenia wykresu
plt.ylabel('Ilość')
plt.xlabel('Wykres')
plt.title('Wojewodztwo')
plt.show()
```

```python
df = data.rodzaj_silnika.value_counts()
df
```

```python
#pobranie wartości 
labels = df.keys().tolist()
#pobreanie ilości wystąpień
sizes = df.tolist()

#utworzenie wykresu liczby występowania aut z konkretnym rodzajem silnika
explode = (0, 0, 0, 0, 0, 0)
plt.pie(sizes, shadow=False, startangle=0,explode=explode, labels=labels, autopct='%1.1f%%')

plt.axis('equal')
plt.title("Diagram kołowy występowania aut z konkretnym rodzajem silnika")
```

```python
#histogram na podstawie przebiegów + 50 krawędzi dla lepszego odczytania
arg  = data['rok_produkcji'].value_counts()
plt.hist(arg, label='Ilości wystąpień lat produkcji', density=True, bins=50)
plt.ylabel('Przebieg')
plt.xlabel('Ilość')
plt.title('Ilość lat produkcji ')
plt.legend(loc="upper right")
plt.show()
```
