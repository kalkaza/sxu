```python
from openpyxl import Workbook
import random
import string
import bs4 as bs
from bs4 import BeautifulSoup
import requests
import urllib.request
```

```python
wb = Workbook()
gielda = wb.create_sheet("Gielda",0)
linki = wb.create_sheet("Linki",1)
filmweb = wb.create_sheet("Filmweb",2)
```

```python
codes = []

def createCode(code_length=3):
    code = ''.join(random.choice(string.ascii_lowercase) for i in range(code_length))
    param = {'s':code}
    link = "http://stooq.pl/q/"
    response = requests.get(link, params=param)
    soup = BeautifulSoup(response.text, "html.parser")
    flag = soup.find(text="Kurs")
    if flag:
        codes.append(code)
        print("Code added: ", code)
    else:
        print("Code not added", code)
```

```python
while len(codes)<5:
    createCode()
print(codes)
```

```python
gielda.cell(row=1,column=1).value = 'Firma'
gielda.cell(row=1,column=2).value = 'Kurs'
gielda.cell(row=1,column=3).value = 'Zmiana'
gielda.cell(row=1,column=4).value = 'Transakcje'

for code in codes:
    param = {'s':code}
    index  = codes.index(code) + 2
    
    link = "http://stooq.pl/q/"
    response = requests.get(link, params=param)
    soup = BeautifulSoup(response.text, "html.parser")
    
    gielda.cell(row=index,column=1).value = code
    gielda.cell(row=index,column=2).value = soup.find(text="Kurs").parent.find("span").text
    gielda.cell(row=index,column=3).value = soup.find(text="Zmiana").parent.find("span").text
    gielda.cell(row=index,column=4).value = soup.find(text="Transakcje").parent.find("span").text
```

```python
response = requests.get("https://stackoverflow.com/")
soup = BeautifulSoup(response.text, 'html.parser')
links = []
for a in soup.find_all("a"):
    links.append(a.get('href'))

for link in links:
    index  = links.index(link) + 1
    if link is None or link.startswith('/') or link.startswith('#'):
        continue
    else:
        linki.cell(row=index,column=1).value = link
        print(link)
print("Links count: ", len(links))
```

```python
response = requests.get("https://www.filmweb.pl/Pulp.Fiction")
soup = BeautifulSoup(response.content, 'html.parser')
#print(soup)
filmweb.cell(row=1,column=1).value = 'Rezyser'
filmweb.cell(row=2,column=1).value = 'Premiera'
filmweb.cell(row=3,column=1).value = 'Boxoffice'
filmweb.cell(row=4,column=1).value = 'Ocena'

direction = filmweb.cell(row=1,column=2).value = soup.find(text="reżyseria").next_element.find("a").text
premiere = filmweb.cell(row=2,column=2).value = soup.find(text="premiera").next_element.find("a").text
boxoffice = soup.find(text="boxoffice").next_element.text
boxoffice = filmweb.cell(row=3,column=2).value = boxoffice.replace('top #27','')
rate = filmweb.cell(row=4,column=2).value = float(soup.find("span", itemprop="ratingValue").text.replace(",", "."))

print(f"ocena: {rate},\nrezyseria: {direction},\npremiera: {premiere},\nboxoffice: {boxoffice}")
```
